```python
import bs4 as bs
import pickle
import requests
import pandas as pd
```

```python
linktopicklefile = input("Input where you want the folder for .pickle file and the .xlsx file")
#/Users/steven/Documents/projects/financeScripts/sp500WikiList
```

```python
def save_sp500_tickers():
    resp = requests.get('http://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
    soup = bs.BeautifulSoup(resp.text, 'lxml')
    table = soup.find('table', {'class': 'wikitable sortable'})
    tickers = []
    for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text
        tickers.append(ticker)

    with open(f"{linktopickle}/p500tickers.pickle","wb") as f:
        pickle.dump(tickers,f)

    df = pd.DataFrame(tickers, columns=["Ticker"])
    df.to_excel(f"{linktopickle}/sp500tickers.xlsx", index=False)
        
    return tickers
```

```python
# save_sp500_tickers()

import datetime
from pandas_datareader import data as pdr

import yfinance as yf
date = datetime.datetime.now().strftime("%Y%m%d%H%M%S")
start = dt.datetime(2023, 11, 27)
end = dt.datetime.now()

def get_data_from_yahoo(reload_sp500=False):
    yf.pdr_override() # <== that's all it takes :-)
    if reload_sp500:
        tickers = save_sp500_tickers()
    else:
        with open(f"{linktopicklefile}/sp500tickers.pickle", "rb") as f:
            tickers = pickle.load(f)
    if os.path.exists(f'{linktosaveData}/{date}/'):
        print('Already have the directory')
    else:
        os.makedirs(f'{linktosaveData}/{date}/')
        
    for ticker in tickers:
        # just in case your connection breaks, we'd like to save our progress!
        print(ticker)
        if not os.path.exists(f'{linktosaveData}/{date}/{ticker}.csv'):
            yf.pdr_override() # <== that's all it takes :-)
            df = pdr.get_data_yahoo(ticker, start=start, end=end)
            df.reset_index(inplace=True)
            df.set_index("Date", inplace=True)
            #df = df.drop("Symbol", axis=1)
            df.to_csv(f'{linktosaveData}/{date}/{ticker}.csv')
        else:
            print('Already have {}'.format(ticker))
```
